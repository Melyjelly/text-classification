```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

res = classifier("i  LOVE HUGGINGFACE") # data we wanna test goes here
print(res)
```

Pipeline: Preprocessing the text with a tokenizer-> feed to the model-> postprocessing( here the label)
