```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from datasets import Dataset
from transformers import AutoTokenizer, pipeline

df = pd.read_csv("../data/test.csv")


#Consists of class ids 1-4 where 1-World, 2-Sports, 3-Business, 4-Sci/Tech



def cleaner(text):
    text = str(text)
    text = text.lower()
    text = re.sub(r"<.*?>", " ", text)
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r"-", " ", text)
    text = re.sub(r"\\", " ", text)
    return text.strip()

df["cleaned"] = df["Description"].map(cleaner)
df.head()





new = df.drop(axis = 1, columns = "Title")


df_train, df_test = train_test_split(new, test_size= 0.25, random_state=42)

print(df_train)


train_dataset = Dataset.from_pandas(df_train)
test_dataset = Dataset.from_pandas(df_test)

print(train_dataset)

tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-uncased")

#tokenized = map(cleaner(), )

```
